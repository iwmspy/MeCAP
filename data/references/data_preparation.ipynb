## 1. Reference split (from the paper written by [Ree et. al.](https://doi.org/10.1039/D4SC07297A))

Since our procedure requires dataset to make the column containing 'train', 'val' and 'test', the integrated split column will be divided to each fold by the following procedure.

```python
# Trial
import pandas as pd
pd.read_csv('./QMdata4ML/df_elec_x.csv.gz',index_col=0,nrows=5)
```

```python
# Make name col from original data

df_origin = pd.read_csv('./QMdata4ML/df_ChEMBL50K.csv.gz',index_col=0)

df_elec = pd.read_csv('./QMdata4ML/df_elec_x.csv.gz',index_col=0)
df_nuc  = pd.read_csv('./QMdata4ML/df_nuc_x.csv.gz',index_col=0)
```

```python
smi_dict = {row['smiles']: row['name'] for idx, row in df_origin.iterrows()}

df_elec['name'] = df_elec['smiles'].map(smi_dict)
df_nuc['name'] = df_nuc['smiles'].map(smi_dict)
```

```python
# Create indivisual 'Fold' columns
def convert_fold(name, fold_num=1):
    if 'train' in name.lower():
        if name.lower() == f'train_fold{fold_num}':
            return 'val'
        else:
            return 'train'
    else:
        return name.lower()

for fnum in range(1,6):
    df_elec[f'Set_fold{fnum}'] = df_elec['Set'].apply(lambda f: convert_fold(f,fnum))
    df_nuc[f'Set_fold{fnum}'] = df_nuc['Set'].apply(lambda f: convert_fold(f,fnum))
```

```python
sf = pd.read_csv('split.csv',index_col=0)['fold'].to_dict()

df_elec['Set_cpbased'] = df_elec['smiles'].map(sf)
df_nuc['Set_cpbased'] = df_nuc['smiles'].map(sf)

for fnum in range(1,6):
    df_elec[f'Set_cpbased_fold{fnum}'] = df_elec['Set_cpbased'].apply(lambda f: convert_fold(f,fnum))
    df_nuc[f'Set_cpbased_fold{fnum}'] = df_nuc['Set_cpbased'].apply(lambda f: convert_fold(f,fnum))
```
