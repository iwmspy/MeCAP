## Reference split (from the paper written by [Ree et. al.](https://doi.org/10.1039/D4SC07297A))

Since our procedure requires dataset to make the column containing 'train', 'val' and 'test', the integrated split column will be divided to each fold by the following procedure.

```python
# Trial
import pandas as pd
pd.read_csv('./QMdata4ML/df_elec_x.csv.gz',index_col=0,nrows=5)
```

```python
# Make name col from original data
df_origin = pd.read_csv('./QMdata4ML/df_ChEMBL50K.csv.gz',index_col=0)

df_elec = pd.read_csv('./QMdata4ML/df_elec_x.csv.gz',index_col=0)
df_nuc  = pd.read_csv('./QMdata4ML/df_nuc_x.csv.gz',index_col=0)

smi_dict = {row['smiles']: row['name'] for idx, row in df_origin.iterrows()}

df_elec['name'] = df_elec['smiles'].map(smi_dict)
df_nuc['name'] = df_nuc['smiles'].map(smi_dict)

# Create indivisual 'Fold' columns
def convert_fold(name, fold_num=1):
    if 'train' in name.lower():
        if name.lower() == f'train_fold{fold_num}':
            return 'val'
        else:
            return 'train'
    else:
        return name.lower()

for fnum in range(1,6):
    df_elec[f'Set_fold{fnum}'] = df_elec['Set'].apply(lambda f: convert_fold(f,fnum))
    df_nuc[f'Set_fold{fnum}'] = df_nuc['Set'].apply(lambda f: convert_fold(f,fnum))
```

## Scaffold-based split

Duplication of compounds is allowed in the reference split. However, the whole information of molecule is input in our model. Therefore, for the guarantee of extraporate, we built the model with Bemis-Murcko-scaffold-splitting.

```python
# !~/miniconda3/envs/mecap/bin/python _ScaffoldSplit.py \
#     --input_csv ./QMdata4ML/df_ChEMBL50K.csv.gz \
#     --smiles_col smiles \
#     --out_split_csv ./split_sim.csv \
#     --out_scaffold_map_csv ./scaf_map.csv \
#     --out_scaffold_pairs_csv ./scaf_pairs.csv \
#     --n_workers 8
```

```python
from tqdm import tqdm
from _rdkitAtomWiseFP import site_feature_vector, add_md_parallel_tqdm

tqdm.pandas()

spl = pd.read_csv('split.csv',index_col=0)

sf = spl['fold'].to_dict()
scaf = spl['scaffold_max_sim_to_train'].to_dict()

df_elec['Set_scaffold'] = df_elec['smiles'].map(sf)
df_nuc['Set_scaffold'] = df_nuc['smiles'].map(sf)

df_nuc['ScafMaxSim'] = df_nuc['smiles'].map(scaf)
df_elec['ScafMaxSim'] = df_elec['smiles'].map(scaf)

df_elec['rdkitAtomWiseFP'] = df_elec.progress_apply(
    lambda row: site_feature_vector(row['smiles'],row['elec_sites']), axis=1)
df_nuc['rdkitAtomWiseFP'] = df_nuc.progress_apply(
    lambda row: site_feature_vector(row['smiles'],row['nuc_sites']), axis=1)

df_elec['MDist'] = add_md_parallel_tqdm(df_elec, 'rdkitAtomWiseFP', 'Set_scaffold', n_workers=10)
df_nuc['MDist'] = add_md_parallel_tqdm(df_nuc, 'rdkitAtomWiseFP', 'Set_scaffold', n_workers=10)
```

## Save split datas

```python
df_elec.to_csv('./QMdata4ML/df_elec_x_with_name_fold.csv')
df_nuc.to_csv('./QMdata4ML/df_nuc_x_with_name_fold.csv')
```
